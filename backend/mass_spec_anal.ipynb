```python
from datetime import datetime
from statistics import mean, stdev
from typing import List
import numpy as np
import matplotlib.pyplot as plt
import csv

def is_starting_index(data_points: List[float], base_line: float, stdev : float):
    minimum = min(data_points)
    if minimum-base_line>stdev:
        return True
    else:
        return False
def subtract_value(list, value):
    # modify the list by subtracting everything
    for i in range(len(list)):
        list[i] -= value

time10 = []
signal10 = []
time20 = []
signal20 = []
time40 = []
signal40 = []
time80 = []
signal80 = []
time160 = []
signal160 = []
time320 = []
signal320 = []

stable_time = 0.2 # min, which has guaranteed non-slope data for any signal before this time point
stable_time_index = 38 # coressponding the time of stable time
polyfit_degree = 5
with open('DBA 20 min.xlsx', newline='') as csvfile:
    spamreader = csv.reader(csvfile)
    for row in spamreader:
        try:
            time10.append(float(row[0]))
            signal10.append(float(row[1]))
            time20.append(float(row[3]))
            signal20.append(float(row[4]))
            time40.append(float(row[6]))
            signal40.append(float(row[7]))
            time80.append(float(row[9]))
            signal80.append(float(row[10]))
            time160.append(float(row[12]))
            signal160.append(float(row[13]))
            time320.append(float(row[15]))
            signal320.append(float(row[16]))
        except:
            pass

plt.plot(time10,signal10)
plt.show()
```

```python
# data for 10 nM
noise_signal10 = signal10[:stable_time_index]
noise_average10 = mean(noise_signal10)
noise_standard_deviation10 = stdev(noise_signal10)
noise_min10 = min(noise_signal10)
noise_max10 = max(noise_signal10)

starting_index10 = stable_time_index # to be modefied

while not is_starting_index(signal10[starting_index10:],noise_average10,noise_standard_deviation10):
    starting_index10 += 1

z10 = np.polyfit(time10[starting_index10:], signal10[starting_index10:], polyfit_degree)
p10 = np.poly1d(z10)
plt.plot(time10[starting_index10:],signal10[starting_index10:], time10[starting_index10:],p10(time10[starting_index10:]))
plt.show()
```

```python
# tailored data
# time all from 0 min,
time10 = [time-time10[starting_index10] for time in time10[starting_index10:]]
time20 = [time-time20[starting_index20] for time in time20[starting_index20:]]
time40 = [time-time40[starting_index40] for time in time40[starting_index40:]]
time80 = [time-time80[starting_index80] for time in time80[starting_index80:]]
time160 = [time-time160[starting_index160] for time in time160[starting_index160:]]
time320 = [time-time320[starting_index320] for time in time320[starting_index320:]]


# signal all from 0
signal10 = [signal - p10(time10)[0] for signal in p10(time10)]
signal20 = [signal - p20(time20)[0] for signal in p20(time20)]
signal40 = [signal - p40(time40)[0] for signal in p40(time40)]
signal80 = [signal - p80(time80)[0] for signal in p80(time80)]
signal160 = [signal - p160(time160)[0] for signal in p160(time160)]
signal320 = [signal - p320(time320)[0] for signal in p320(time320)]
plt.plot(time10,signal10,
         time20,signal20,
         time40,signal40,
         time80,signal80,
         time160,signal160,
         time320,signal320)
plt.legend(['10 nM','20 nM','40 nM','80 nM','160 nM','320 nM'])

plt.show()
```
